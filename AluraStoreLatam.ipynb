### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: {ingreso_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_tienda4}")
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

# Truncar a un decimal sin redondear
calificacion_formato_tienda1 = int(calificacion_promedio_tienda1 * 100) / 100
calificacion_formato_tienda2 = int(calificacion_promedio_tienda2 * 100) / 100
calificacion_formato_tienda3 = int(calificacion_promedio_tienda3 * 100) / 100
calificacion_formato_tienda4 = int(calificacion_promedio_tienda4 * 100) / 100


print(f"Calificación promedio Tienda 1: {calificacion_formato_tienda1}")
print(f"Calificación promedio Tienda 2: {calificacion_formato_tienda2}")
print(f"Calificación promedio Tienda 3: {calificacion_formato_tienda3}")
print(f"Calificación promedio Tienda 4: {calificacion_formato_tienda4}")
```

# Distribución Geográfica de las **Ventas**

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import matplotlib.pyplot as plt # Para gráficos de dispersión si lo prefieres
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# Asegúrate de que tus DataFrames (tienda, tienda2, tienda3, tienda4) estén cargados
# Puedes usar el código de carga que ya tienes.

# Prepara los datos para el heatmap de cada tienda (lista de [lat, lon])
datos_heatmap_tienda1 = tienda[['lat', 'lon']].values.tolist()
datos_heatmap_tienda2 = tienda2[['lat', 'lon']].values.tolist()
datos_heatmap_tienda3 = tienda3[['lat', 'lon']].values.tolist()
datos_heatmap_tienda4 = tienda4[['lat', 'lon']].values.tolist()

# Calcula el centroide aproximado de todas las tiendas para centrar el mapa (opcional)
centro_lat = (tienda['lat'].mean() + tienda2['lat'].mean() + tienda3['lat'].mean() + tienda4['lat'].mean()) / 4
centro_lon = (tienda['lon'].mean() + tienda2['lon'].mean() + tienda3['lon'].mean() + tienda4['lon'].mean()) / 4

# Crea un mapa base centrado
mapa_ventas_heatmap = folium.Map(location=[centro_lat, centro_lon], zoom_start=10) # Ajusta el zoom según sea necesario

# Agrega una capa de HeatMap para cada tienda con un gradiente de color diferente
# Puedes ajustar el radio y los gradientes de color según prefieras
HeatMap(datos_heatmap_tienda1, radius=15, gradient={0.4: 'blue', 0.6: 'cyan', 0.8: 'lightblue'}).add_to(mapa_ventas_heatmap)
HeatMap(datos_heatmap_tienda2, radius=15, gradient={0.4: 'green', 0.6: 'lightgreen', 0.8: 'lime'}).add_to(mapa_ventas_heatmap)
HeatMap(datos_heatmap_tienda3, radius=15, gradient={0.4: 'darkred', 0.6: 'red', 0.8: 'salmon'}).add_to(mapa_ventas_heatmap)
HeatMap(datos_heatmap_tienda4, radius=15, gradient={0.4: 'darkorange', 0.6: 'orange', 0.8: 'gold'}).add_to(mapa_ventas_heatmap)


# Para que las diferentes capas de heatmap sean visibles y se puedan activar/desactivar
# puedes agregarlas a un Layer Control.
folium.LayerControl().add_to(mapa_ventas_heatmap)


# Muestra el mapa
mapa_ventas_heatmap
```
